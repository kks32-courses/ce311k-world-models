## üß† What Are Conditional Statements? (The Engineering "If-Then" Logic)

**Conditional statements** are how we teach computers to make decisions, just like engineers do:

**Engineering Example:**
- *IF* the traffic incident happened in 2020, *THEN* it's during the pandemic
- *IF* the incident type is "Crash Urgent", *THEN* it's a high-priority safety issue
- *IF* the incident is on IH-35, *THEN* it affects the main Austin traffic corridor

**Why Do We Need This?**
- We have hundreds of thousands of incidents across multiple years
- We need to separate pre-pandemic (2019) from pandemic (2020) data
- We want to compare patterns and find the hidden story

Let's learn conditional logic by investigating Austin's pandemic traffic story!

## üìÖ Working with Dates (Time-Based Engineering Analysis)

Before we can compare years, we need to extract the year from our date information. This is a common challenge in engineering data analysis.

**The Challenge:**
- Our dates are stored as text: "03/06/2024 01:29:39 AM +0000"
- We need to extract just the year: 2024
- Then we can use conditionals to filter by year

**Engineering Application:**
- Compare traffic patterns before, during, and after the pandemic
- Identify seasonal trends
- Analyze peak incident times

```python
print("\n\n=== SESSION 2: Finding the Pandemic Traffic Signal ===")

# TOPIC: Working with dates for time-based analysis
print("\n--- Preparing Data for Time Analysis ---")

# Convert date strings to datetime objects (so Python understands them as dates)
traffic_data['Date'] = pd.to_datetime(traffic_data['Published Date'])
traffic_data['Year'] = traffic_data['Date'].dt.year
traffic_data['Month'] = traffic_data['Date'].dt.month
traffic_data['Hour'] = traffic_data['Date'].dt.hour

print("Added time information to help analyze pandemic effects:")
print(traffic_data[['Published Date', 'Year', 'Month', 'Hour']].head(3))

print(f"\nYears in our dataset: {sorted(traffic_data['Year'].unique())}")
```

## ‚úÖ Boolean Logic (True/False Engineering Decisions)

**Boolean** values are simply True or False - perfect for engineering yes/no decisions.

**Traffic Engineering Examples:**
- Is this incident during the pandemic? (True/False)
- Is this a crash? (True/False)  
- Is this location on a major highway? (True/False)

**Comparison Operators** (how we create True/False conditions):
- `==` : equal to
- `!=` : not equal to  
- `>` : greater than
- `<` : less than
- `>=` : greater than or equal
- `<=` : less than or equal

Let's use these to filter our traffic data!

```python
# TOPIC: Boolean Logic and Conditional Statements
print("\n--- Understanding Boolean Logic Through Traffic Data ---")

# Create boolean conditions (True/False for each incident)
is_2019 = traffic_data['Year'] == 2019  # Pre-pandemic
is_2020 = traffic_data['Year'] == 2020  # Pandemic year
is_2021 = traffic_data['Year'] == 2021  # Post-pandemic

print("Boolean conditions created:")
print(f"is_2019 (first 5 values): {is_2019.head().tolist()}")
print(f"is_2020 (first 5 values): {is_2020.head().tolist()}")

# Count True values (how many incidents per year)
count_2019 = is_2019.sum()  # sum() counts True values!
count_2020 = is_2020.sum()
count_2021 = is_2021.sum()

print(f"\nIncident counts by year:")
print(f"2019 (Pre-pandemic): {count_2019:,} incidents")
print(f"2020 (Pandemic year): {count_2020:,} incidents")
print(f"2021 (Post-pandemic): {count_2021:,} incidents")
```

## üîç Data Filtering (Finding the Pandemic Signal)

Now we'll use our boolean conditions to **filter** our data - one of the most powerful techniques in data analysis!

**How Filtering Works:**
1. Create a boolean condition (True/False for each row)
2. Use that condition to select only the rows where it's True
3. Analyze the filtered subset

**Engineering Application:**
- Filter to only 2019 data to establish our "normal" baseline
- Filter to only 2020 data to see pandemic effects
- Compare the two to find the story

```python
# TOPIC: Data Filtering with Conditional Statements
print("\n--- Using Conditional Statements to Filter by Year ---")

# Apply filters to get subsets of data
pre_pandemic_data = traffic_data[is_2019]    # Only 2019 data
pandemic_data = traffic_data[is_2020]        # Only 2020 data  
post_pandemic_data = traffic_data[is_2021]   # Only 2021 data

print(f"Filtered datasets created:")
print(f"Pre-pandemic (2019): {len(pre_pandemic_data):,} incidents")
print(f"Pandemic year (2020): {len(pandemic_data):,} incidents")
print(f"Post-pandemic (2021): {len(post_pandemic_data):,} incidents")

# Calculate the change
if len(pre_pandemic_data) > 0:
    pandemic_change = len(pandemic_data) - len(pre_pandemic_data)
    percent_change = (pandemic_change / len(pre_pandemic_data)) * 100
    
    print(f"\nüö® PANDEMIC IMPACT ANALYSIS:")
    print(f"Change in incidents (2019 ‚Üí 2020): {pandemic_change:+,}")
    print(f"Percent change: {percent_change:+.1f}%")
    
    if percent_change < -20:
        print("üìâ Traffic incidents DECREASED significantly during the pandemic!")
    elif percent_change > 20:
        print("üìà Traffic incidents INCREASED significantly during the pandemic!")
    else:
        print("‚û°Ô∏è  Traffic incidents remained relatively stable.")
else:
    print("‚ö†Ô∏è  Not enough 2019 data for comparison")
```

## üîÑ Complex Conditional Logic (Multiple Conditions)

Real engineering problems often require multiple conditions at once:

**Boolean Operators:**
- `&` : AND (both conditions must be True)
- `|` : OR (either condition can be True)  
- `~` : NOT (opposite of the condition)

**Engineering Examples:**
- Find incidents that are both in March AND in 2020
- Find incidents that are either crashes OR stalled vehicles
- Find incidents that are NOT during rush hour

Let's use this to dig deeper into the pandemic story!

```python
# TOPIC: Complex Filtering with Multiple Conditions
print("\n--- More Complex Filtering: March 2019 vs March 2020 ---")

# Compare the same month across years (control for seasonal effects)
march_2019 = traffic_data[(traffic_data['Year'] == 2019) & (traffic_data['Month'] == 3)]
march_2020 = traffic_data[(traffic_data['Year'] == 2020) & (traffic_data['Month'] == 3)]

print(f"March 2019 incidents: {len(march_2019):,}")
print(f"March 2020 incidents: {len(march_2020):,}")

# Calculate the change (this controls for seasonal variations)
if len(march_2019) > 0:
    march_change = len(march_2020) - len(march_2019)
    march_percent_change = (march_change / len(march_2019)) * 100
    
    print(f"\nüìä MARCH-TO-MARCH COMPARISON:")
    print(f"Change in incidents (Mar 2019 ‚Üí Mar 2020): {march_change:+,}")
    print(f"Percent change: {march_percent_change:+.1f}%")
    
    print(f"\nüí° ENGINEERING INSIGHT:")
    print(f"Comparing the same month eliminates seasonal effects!")
    print(f"This gives us a cleaner signal of pandemic impact.")
else:
    print("‚ö†Ô∏è  Not enough March 2019 data for comparison")
```

## üìà Data Visualization (Making the Story Visible)

Numbers tell us what happened, but **visualizations** help us see the story clearly.

**Why Engineers Use Plots:**
- Patterns jump out visually that are hidden in tables
- Easy to communicate findings to stakeholders
- Helps validate our analysis ("Does this make sense?")
- Essential for engineering reports and presentations

Let's create visualizations that reveal Austin's pandemic traffic story!

```python
# TOPIC: Basic Plotting to Visualize the Traffic Story
print("\n--- Plotting Traffic Incidents to See the Pandemic Story ---")

# Count incidents by year
yearly_counts = traffic_data['Year'].value_counts().sort_index()

# Create our first plot
plt.figure(figsize=(12, 8))

# Plot 1: Bar chart of incidents by year
plt.subplot(2, 2, 1)
years = yearly_counts.index
counts = yearly_counts.values
colors = ['blue' if year < 2020 else 'red' if year == 2020 else 'green' for year in years]
bars = plt.bar(years, counts, color=colors)
plt.title('Austin Traffic Incidents by Year\n(Pandemic Impact)', fontsize=12, fontweight='bold')
plt.xlabel('Year')
plt.ylabel('Number of Incidents')
plt.xticks(years)

# Add value labels on bars
for bar, value in zip(bars, counts):
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 500, 
             f'{value:,}', ha='center', va='bottom', fontweight='bold')

# Plot 2: Pie chart of incident types
plt.subplot(2, 2, 2)
incident_type_counts = traffic_data['Issue Reported'].value_counts().head(4)
plt.pie(incident_type_counts.values, labels=incident_type_counts.index, autopct='%1.1f%%')
plt.title('Types of Traffic Incidents\n(What Problems Dominate?)', fontsize=12, fontweight='bold')

plt.tight_layout()
plt.show()

print("\nüéØ What do you notice in these visualizations?")
print("üí≠ Think like an engineer: What questions do these patterns raise?")
```

## ‚è∞ Time-Based Analysis (When Do Incidents Happen?)

Let's dig deeper into the temporal patterns. Did the pandemic change WHEN incidents happen during the day?

**Engineering Questions:**
- Did rush hour patterns change during the pandemic?
- Are there different peak times in 2020 vs 2019?
- What does this tell us about Austin's work patterns?

This type of analysis helps transportation engineers plan for the future!

```python
# TOPIC: Time-based Analysis - When Do Incidents Happen?
print("\n--- When Do Traffic Incidents Happen? ---")

# Compare hourly patterns for 2019 vs 2020
plt.figure(figsize=(14, 6))

# Get hourly data for each year
hours = range(24)
hourly_2019 = [len(pre_pandemic_data[pre_pandemic_data['Hour'] == h]) for h in hours]
hourly_2020 = [len(pandemic_data[pandemic_data['Hour'] == h]) for h in hours]

# Plot comparison
plt.plot(hours, hourly_2019, 'b-o', linewidth=2, label='2019 (Pre-Pandemic)')
plt.plot(hours, hourly_2020, 'r-o', linewidth=2, label='2020 (Pandemic)')

plt.title('Traffic Incidents by Hour: Pandemic Impact\n(Did Rush Hour Change?)', 
          fontsize=14, fontweight='bold')
plt.xlabel('Hour of Day (24-hour format)')
plt.ylabel('Number of Incidents')
plt.legend()
plt.grid(True, alpha=0.3)
plt.xticks(range(0, 24, 2))

# Add rush hour indicators
plt.axvline(x=8, color='gray', linestyle='--', alpha=0.5)
plt.axvline(x=17, color='gray', linestyle='--', alpha=0.5)
plt.text(8, max(max(hourly_2019), max(hourly_2020))*0.9, 'Morning\nRush', 
         ha='center', fontsize=10, alpha=0.7)
plt.text(17, max(max(hourly_2019), max(hourly_2020))*0.9, 'Evening\nRush', 
         ha='center', fontsize=10, alpha=0.7)

plt.tight_layout()
plt.show()

# Find peak hours
peak_2019 = hours[hourly_2019.index(max(hourly_2019))]
peak_2020 = hours[hourly_2020.index(max(hourly_2020))]

print(f"\nüïê PEAK INCIDENT HOURS:")
print(f"2019: {peak_2019}:00 ({max(hourly_2019):,} incidents)")
print(f"2020: {peak_2020}:00 ({max(hourly_2020):,} incidents)")
```

## üîÑ Why Do Engineers Need Loops? (Automating Repetitive Analysis)

**The Problem:** In Week 1, we manually analyzed 2019, 2020, and 2021 data separately. What if we had 10 years of data? Or wanted to analyze every Austin location? We'd be copy-pasting code forever!

**The Solution:** **Loops** let us automate repetitive tasks - one of the most powerful concepts in computational engineering.

**Real Engineering Applications:**
- Analyze traffic patterns for ALL years in our dataset
- Check incident trends at EVERY major Austin intersection
- Calculate statistics for EACH type of incident
- Process thousands of data points in seconds instead of hours

**Think Like an Engineer:** If you find yourself doing the same analysis over and over, there's probably a loop that can automate it!

Let's see loops in action with our Austin traffic data.

```python
print("\n\n=== WEEK 2: AUTOMATING ANALYSIS WITH LOOPS ===")

# TOPIC: Basic For Loop Concept
print("\n--- Understanding For Loops Through Traffic Analysis ---")

# Define years to analyze
years_to_analyze = [2019, 2020, 2021]
print(f"Years selected for pandemic analysis: {years_to_analyze}")

print("\nUsing a for loop to process each year:")
for year in years_to_analyze:
    year_data = traffic_data[traffic_data['Year'] == year]
    incident_count = len(year_data)
    print(f"  Year {year}: {incident_count:,} incidents")

print("\nüí° ENGINEERING INSIGHT:")
print("   Instead of writing 3 separate analyses, we wrote 1 loop!")
print("   This scales to ANY number of years automatically.")
```

- **Data Organization**: Structure results for easy interpretation

Let's create a comprehensive analysis that an Austin transportation engineer could actually use!

```python
# TOPIC: Advanced Loop Analysis with Data Storage
print("\n--- Automated Incident Analysis by Year ---")

# Dictionary to store results (professional data organization)
yearly_analysis = {}

for year in years_to_analyze:
    print(f"\nAnalyzing {year}...")
    
    # Filter data for this year
    year_data = traffic_data[traffic_data['Year'] == year]
    
    # Calculate various statistics
    total_incidents = len(year_data)
    
    # Most common incident type and location
    if len(year_data) > 0:
        most_common_incident = year_data['Issue Reported'].value_counts().index[0]
        most_problematic_location = year_data['Address'].value_counts().index[0]
    else:
        most_common_incident = "None"
        most_problematic_location = "None"
    
    # Store results in organized format
    yearly_analysis[year] = {
        'total_incidents': total_incidents,
        'most_common_incident': most_common_incident,
        'most_problematic_location': most_problematic_location
    }
    
    # Print summary for this year
    print(f"  Total incidents: {total_incidents:,}")
    print(f"  Most common incident: {most_common_incident}")
    print(f"  Most problematic location: {most_problematic_location[:50]}...")

print("\nüéØ ANALYSIS COMPLETE! Results stored for engineering report.")
```

```python
# TOPIC: Converting Results to Professional Format
print("\n--- Creating Summary Table from Our Analysis ---")

# Convert our analysis to a DataFrame for better visualization
summary_df = pd.DataFrame(yearly_analysis).transpose()
print("\nüìã PROFESSIONAL ENGINEERING SUMMARY:")
print(summary_df)

# Calculate pandemic impact metrics
if 2019 in yearly_analysis and 2020 in yearly_analysis:
    incidents_2019 = yearly_analysis[2019]['total_incidents']
    incidents_2020 = yearly_analysis[2020]['total_incidents']
    
    if incidents_2019 > 0:
        pandemic_impact = ((incidents_2020 - incidents_2019) / incidents_2019) * 100
        print(f"\nüìä PANDEMIC IMPACT CALCULATION:")
        print(f"   2019 Baseline: {incidents_2019:,} incidents")
        print(f"   2020 Pandemic: {incidents_2020:,} incidents")
        print(f"   Change: {pandemic_impact:+.1f}%")
        
        if pandemic_impact < -20:
            print(f"   üö® MAJOR DECREASE: Significant reduction in traffic incidents")
        elif pandemic_impact > 20:
            print(f"   üö® MAJOR INCREASE: Significant rise in traffic incidents")
        else:
            print(f"   ‚ÑπÔ∏è  MODERATE CHANGE: Relatively stable patterns")
```

## üó∫Ô∏è Geographic Analysis with Loops (Mapping Austin's Traffic Hotspots)

Now let's use loops to analyze traffic patterns across Austin's geography. This is exactly the type of analysis transportation engineers do to prioritize infrastructure improvements!

**Engineering Questions:**
- Which Austin locations have the most incidents?
- How did different areas respond to the pandemic?
- Where should the city focus improvement efforts?

**Computational Challenge:**
- Austin has hundreds of different incident locations
- Manual analysis would take days
- Loops can process them all in seconds!

```python
# TOPIC: Geographic Analysis with Loops
print("\n--- Analyzing Austin Locations Automatically ---")

# Get unique locations
unique_locations = traffic_data['Address'].dropna().unique()
print(f"Found {len(unique_locations):,} unique Austin locations in our dataset")

# Analyze top locations
location_counts = traffic_data['Address'].value_counts().head(10)
print(f"\nüèôÔ∏è TOP 10 AUSTIN TRAFFIC HOTSPOTS:")
for i, (location, count) in enumerate(location_counts.items(), 1):
    print(f"   {i:2d}. {location}: {count:,} incidents")

# Analyze pandemic impact by location
location_analysis = {}
top_locations = location_counts.head(5).index

print(f"\n--- Detailed Pandemic Analysis for Top 5 Locations ---")

for location in top_locations:
    print(f"\nüìç Analyzing: {location}")
    
    # Filter data for this location
    location_data = traffic_data[traffic_data['Address'] == location]
    
    # Analyze by year for this location
    incidents_by_year = {}
    for year in years_to_analyze:
        year_location_data = location_data[location_data['Year'] == year]
        incidents_by_year[year] = len(year_location_data)
    
    location_analysis[location] = incidents_by_year
    
    # Show the trend
    print(f"   2019: {incidents_by_year[2019]:,} incidents")
    print(f"   2020: {incidents_by_year[2020]:,} incidents")
    print(f"   2021: {incidents_by_year[2021]:,} incidents")
    
    # Calculate pandemic change
    if incidents_by_year[2019] > 0:
        pandemic_change = ((incidents_by_year[2020] - incidents_by_year[2019]) / incidents_by_year[2019]) * 100
        print(f"   üìä Pandemic change: {pandemic_change:+.1f}%")
        
        if pandemic_change < -30:
            print(f"   ‚úÖ MAJOR IMPROVEMENT: Large reduction in incidents")
        elif pandemic_change > 30:
            print(f"   ‚ö†Ô∏è  INCREASED RISK: Large increase in incidents")
        else:
            print(f"   ‚ÑπÔ∏è  STABLE: Moderate change in incident patterns")
```

## üìä Histograms and Distribution Analysis (Understanding Traffic Patterns)

**Histograms** are perfect for understanding the distribution of engineering data - they show us HOW incidents are spread across time, space, and type.

```python
# TOPIC: Comprehensive Histogram Analysis
print("\n--- Using Histograms to Understand Austin Traffic Patterns ---")

# Create comprehensive histogram analysis
fig, axes = plt.subplots(2, 2, figsize=(16, 12))
fig.suptitle('Austin Traffic Incident Distributions: Complete Engineering Analysis', 
             fontsize=16, fontweight='bold')

# Histogram 1: Incidents by hour of day
axes[0, 0].hist(traffic_data['Hour'].dropna(), bins=24, alpha=0.7, color='skyblue', edgecolor='black')
axes[0, 0].set_title('Daily Traffic Incident Patterns\n(When Do Incidents Occur?)', fontweight='bold')
axes[0, 0].set_xlabel('Hour of Day (24-hour format)')
axes[0, 0].set_ylabel('Number of Incidents')
axes[0, 0].set_xticks(range(0, 24, 2))
axes[0, 0].axvline(x=8, color='red', linestyle='--', alpha=0.7, label='Morning Rush')
axes[0, 0].axvline(x=17, color='red', linestyle='--', alpha=0.7, label='Evening Rush')
axes[0, 0].legend()

# Histogram 2: Incidents by year
yearly_counts = traffic_data['Year'].value_counts().sort_index()
colors_by_year = ['lightblue' if year < 2020 else 'red' if year == 2020 else 'lightgreen' 
                  for year in yearly_counts.index]
bars = axes[0, 1].bar(yearly_counts.index, yearly_counts.values, 
                     color=colors_by_year, alpha=0.7, edgecolor='black')
axes[0, 1].set_title('Austin Traffic Incidents by Year\n(Pandemic Impact Visualization)', fontweight='bold')
axes[0, 1].set_xlabel('Year')
axes[0, 1].set_ylabel('Number of Incidents')
axes[0, 1].tick_params(axis='x', rotation=45)

for bar, value in zip(bars, yearly_counts.values):
    axes[0, 1].text(bar.get_x() + bar.get_width()/2, bar.get_height() + 1000, 
                   f'{value:,}', ha='center', va='bottom', fontweight='bold', fontsize=9)

# Histogram 3: Top incident types
incident_type_counts = traffic_data['Issue Reported'].value_counts().head(8)
bars3 = axes[1, 0].barh(range(len(incident_type_counts)), incident_type_counts.values, 
                       color='orange', alpha=0.7)
axes[1, 0].set_yticks(range(len(incident_type_counts)))
axes[1, 0].set_yticklabels(incident_type_counts.index, fontsize=10)
axes[1, 0].set_title('Types of Traffic Incidents in Austin\n(Engineering Priority Analysis)', fontweight='bold')
axes[1, 0].set_xlabel('Number of Incidents')

for i, (bar, value) in enumerate(zip(bars3, incident_type_counts.values)):
    axes[1, 0].text(value + 1000, i, f'{value:,}', va='center', fontweight='bold', fontsize=9)

# Histogram 4: Seasonal patterns
axes[1, 1].hist(traffic_data['Month'].dropna(), bins=12, alpha=0.7, color='lightgreen', edgecolor='black')
axes[1, 1].set_title('Seasonal Traffic Incident Patterns\n(Monthly Distribution)', fontweight='bold')
axes[1, 1].set_xlabel('Month')
axes[1, 1].set_ylabel('Number of Incidents')
axes[1, 1].set_xticks(range(1, 13))
axes[1, 1].set_xticklabels(['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun',
                           'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'], rotation=45)

plt.tight_layout()
plt.show()

print("\nüéØ HISTOGRAM INSIGHTS:")
print("   ‚Ä¢ Visual patterns reveal engineering opportunities")
print("   ‚Ä¢ Multiple dimensions of analysis in one comprehensive view")
print("   ‚Ä¢ Clear identification of peak times, seasons, and incident types")
print("   ‚Ä¢ Data-driven foundation for infrastructure planning")
```

Now let's create the ultimate visualization - a comprehensive analysis that tells Austin's complete traffic story. This combines everything we've learned: loops, filtering, and advanced plotting.

**Engineering Goal:** Create a professional analysis that Austin's transportation department could actually use for infrastructure planning and policy decisions.

```python
# TOPIC: Comprehensive Austin Traffic Story Visualization
print("\n--- Creating Austin's Complete Traffic Analysis Dashboard ---")

# Create the ultimate traffic analysis visualization
fig, axes = plt.subplots(2, 2, figsize=(18, 14))
fig.suptitle('AUSTIN TRAFFIC PANDEMIC STORY: Complete Engineering Analysis\n' + 
             'Data-Driven Insights for Transportation Planning', 
             fontsize=18, fontweight='bold', y=0.98)

# Plot 1: Top locations with pandemic impact
top_locations_data = traffic_data['Address'].value_counts().head(8)
colors = ['red' if any(highway in str(loc).upper() for highway in ['IH 35', 'I-35']) else 
          'blue' if 'MOPAC' in str(loc).upper() or 'MO PAC' in str(loc).upper() else 
          'green' for loc in top_locations_data.index]

bars1 = axes[0, 0].barh(range(len(top_locations_data)), top_locations_data.values, color=colors, alpha=0.7)
axes[0, 0].set_yticks(range(len(top_locations_data)))
short_labels = [str(loc)[:40] + '...' if len(str(loc)) > 40 else str(loc) for loc in top_locations_data.index]
axes[0, 0].set_yticklabels(short_labels, fontsize=9)
axes[0, 0].set_title('Austin Traffic Hotspots\n(Total Incidents by Location)', fontweight='bold')
axes[0, 0].set_xlabel('Number of Incidents')

for i, (bar, value) in enumerate(zip(bars1, top_locations_data.values)):
    axes[0, 0].text(value + 50, i, f'{value:,}', va='center', fontweight='bold', fontsize=9)

# Plot 2: Pandemic impact timeline
pandemic_years = [2019, 2020, 2021]
pandemic_data_counts = [len(traffic_data[traffic_data['Year'] == year]) for year in pandemic_years]
colors2 = ['blue', 'red', 'green']

bars2 = axes[0, 1].bar(pandemic_years, pandemic_data_counts, color=colors2, alpha=0.7, width=0.6)
axes[0, 1].set_title('Pandemic Impact Timeline\n(2019 Baseline ‚Üí 2020 Pandemic ‚Üí 2021 Recovery)', fontweight='bold')
axes[0, 1].set_xlabel('Year')
axes[0, 1].set_ylabel('Number of Incidents')
axes[0, 1].set_xticks(pandemic_years)

for i, (bar, count) in enumerate(zip(bars2, pandemic_data_counts)):
    axes[0, 1].text(bar.get_x() + bar.get_width()/2, bar.get_height() + 1000, 
                   f'{count:,}', ha='center', va='bottom', fontweight='bold')
    if i > 0:
        prev_count = pandemic_data_counts[i-1]
        if prev_count > 0:
            change_pct = ((count - prev_count) / prev_count) * 100
            axes[0, 1].text(bar.get_x() + bar.get_width()/2, bar.get_height()/2, 
                           f'{change_pct:+.0f}%', ha='center', va='center', 
                           fontweight='bold', color='white', fontsize=12)

# Plot 3: Incident severity analysis
severity_categories = {
    'HIGH PRIORITY': ['Crash Urgent', 'COLLISION WITH INJURY', 'TRAFFIC FATALITY', 'AUTO/ PED'],
    'MEDIUM PRIORITY': ['Crash Service', 'COLLISION', 'VEHICLE FIRE'],
    'LOW PRIORITY': ['Traffic Hazard', 'TRFC HAZD/ DEBRIS', 'Stalled Vehicle']
}

severity_counts = {}
for category, incident_types in severity_categories.items():
    count = sum(len(traffic_data[traffic_data['Issue Reported'] == incident_type]) 
                for incident_type in incident_types 
                if incident_type in traffic_data['Issue Reported'].values)
    severity_counts[category] = count

labels = list(severity_counts.keys())
sizes = list(severity_counts.values())
colors3 = ['red', 'orange', 'yellow']
explode = (0.1, 0, 0)

wedges, texts, autotexts = axes[1, 0].pie(sizes, labels=labels, colors=colors3, autopct='%1.1f%%',
                                         explode=explode, shadow=True, startangle=90)
axes[1, 0].set_title('Incident Severity Distribution\n(Engineering Priority Classification)', fontweight='bold')

for autotext in autotexts:
    autotext.set_fontweight('bold')
    autotext.set_fontsize(11)

# Plot 4: Rush hour analysis with pandemic comparison
hours = range(24)
rush_hour_data = {}

for year in [2019, 2020, 2021]:
    year_data = traffic_data[traffic_data['Year'] == year]
    hourly_counts = [len(year_data[year_data['Hour'] == hour]) for hour in hours]
    rush_hour_data[year] = hourly_counts

colors4 = ['blue', 'red', 'green']
labels4 = ['2019 (Pre-Pandemic)', '2020 (Pandemic)', '2021 (Recovery)']

for i, (year, data) in enumerate(rush_hour_data.items()):
    axes[1, 1].plot(hours, data, marker='o', linewidth=2, color=colors4[i], 
                   label=labels4[i], markersize=4)

axes[1, 1].set_title('Rush Hour Pattern Analysis\n(How Pandemic Changed Daily Traffic)', fontweight='bold')
axes[1, 1].set_xlabel('Hour of Day (24-hour format)')
axes[1, 1].set_ylabel('Number of Incidents')
axes[1, 1].legend(loc='upper right')
axes[1, 1].grid(True, alpha=0.3)
axes[1, 1].set_xticks(range(0, 24, 2))
axes[1, 1].axvspan(7, 9, alpha=0.2, color='red')
axes[1, 1].axvspan(16, 18, alpha=0.2, color='red')

plt.tight_layout()
plt.show()

print("\nüéØ COMPREHENSIVE ANALYSIS COMPLETE!")
print("   This dashboard provides actionable insights for Austin transportation planning.")
```
